```python
Dat2016 = pd.merge(Data2016,GeoRef,how='left',left_on=Data2016['fips_code'],right_on=GeoRef["county_fips"])
Dat2016.sort_values(by=['fips_code'])
Dat2016['gop_votes'] = Dat2016['gop_2016']
Dat2016['dem_votes'] = Dat2016['dem_2016']
Dat2016['gop_dem_total'] = Dat2016['total_2016']
Dat2016['county_name'] = Dat2016['county']


Dat2016 = Dat2016[['state_name','fips_code','county_name','gop_votes','dem_votes','gop_dem_total']]
Dat2016['year'] = 2016
```

```python
Dat2012 = pd.merge(Data2012,GeoRef,how='left',left_on=Data2012['fips_code'],right_on=GeoRef["county_fips"])
Dat2012.sort_values(by=['fips_code'])
Dat2012['gop_votes'] = Dat2012['gop_2012']
Dat2012['dem_votes'] = Dat2012['dem_2012']
Dat2012['gop_dem_total'] = Dat2012['total_2012']
Dat2012['county_name'] = Dat2012['county']


Dat2012 = Dat2012[['state_name','fips_code','county_name','gop_votes','dem_votes','gop_dem_total']]
Dat2012['year'] = 2012
```

```python
Dat2008 = pd.merge(Data2008,GeoRef,how='left',left_on=Data2008['fips_code'],right_on=GeoRef["county_fips"])
Dat2008.sort_values(by=['fips_code'])

Dat2008['gop_votes'] = Dat2008['gop_2008']
Dat2008['dem_votes'] = Dat2008['dem_2008']
Dat2008['gop_dem_total'] = Dat2008['total_2008']
Dat2008['county_name'] = Dat2008['county']


Dat2008 = Dat2008[['state_name','fips_code','county_name','gop_votes','dem_votes','gop_dem_total']]
Dat2008['year'] = 2008
Dat2008.loc[Dat2008.gop_dem_total == 2024, "gop_dem_total"] = 16323
```

```python
data08thru16 = pd.concat([Dat2016,Dat2012,Dat2008])
data08thru16['margin'] = (data08thru16['gop_votes'] - data08thru16['dem_votes']).abs()
data08thru16['per_gop'] = data08thru16['gop_votes'] / data08thru16['gop_dem_total']
data08thru16['per_dem'] = data08thru16['dem_votes'] / data08thru16['gop_dem_total']
data08thru16['perc_margin'] = (data08thru16['per_gop'] - data08thru16['per_dem']).abs()
data08thru16 = data08thru16.sort_values(by=['year','fips_code'])
data08thru16.head(1)
```
